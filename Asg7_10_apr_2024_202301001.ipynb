```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Feature (independent variable)
y = np.array([2, 3, 4, 8, 7])                # Target (dependent variable)

# Create linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the fitted line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Print the slope (coefficient) and intercept of the fitted line
print('Slope:', model.coef_[0])
print('Intercept:', model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 7, 5])

# Number of observations
n = len(x)

# Means of x and y
m_x = np.mean(x)
m_y = np.mean(y)

# Calculate cross-deviation and deviation about x
SS_xy = np.sum(y * x) - n * m_y * m_x
SS_xx = np.sum(x * x) - n * m_x * m_x

# Regression coefficients
b_1 = SS_xy / SS_xx
b_0 = m_y - b_1 * m_x

# Print the coefficients
print("Slope (b_1):", b_1)
print("Intercept (b_0):", b_0)

# Predicted response vector
y_pred = b_0 + b_1 * x

plt.scatter(x, y, color="m", marker="o", s=30)
plt.plot(x, y_pred, color="g")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
