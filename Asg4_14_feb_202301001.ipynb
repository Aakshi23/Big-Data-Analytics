```python
import pandas as pd
import matplotlib.pyplot as plt
url="/content/drive/MyDrive/company_sales_data.csv"
data=pd.read_csv(url)
print (data)

plt.plot(data['month_number'],data['total_profit'])
plt.title("Line Chart")

plt.xlabel('month')
plt.ylabel('total')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
url="/content/drive/MyDrive/company_sales_data.csv"
data=pd.read_csv(url)
print (data)

plt.plot(data['month_number'])
plt.plot(data['facecream'])
plt.plot(data['facewash'])
plt.plot(data['toothpaste'])
plt.plot(data['bathingsoap'])
plt.plot(data['shampoo'])
plt.plot(data['moisturizer'])
plt.title("Line Chart")

plt.xlabel('month')
plt.ylabel('total')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url="/content/drive/MyDrive/company_sales_data.csv"
data=pd.read_csv(url)
print (data)

plt.bar(data['month_number'],data['total_profit'])
plt.title("Bar Chart")

plt.xlabel('month')
plt.ylabel('total')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url="/content/drive/MyDrive/company_sales_data.csv"
data=pd.read_csv(url)
print (data)

plt.scatter(data['month_number'],data['facecream'],data['facewash'],data['shampoo'])
plt.title("Scatter Chart")

plt.xlabel('month')
plt.ylabel('total')

plt.colorbar()

plt.show()
```
